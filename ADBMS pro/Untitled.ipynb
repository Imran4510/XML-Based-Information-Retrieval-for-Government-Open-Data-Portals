```python
import pandas as pd

df= pd.read_csv("ICRISAT-District Level Data.csv")

df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df.rename(columns={
    "State Name": "state",
    "Dist Name": "district",
    "Year": "year",
    
    "RICE AREA (1000 ha)": "rice_area",
    "RICE PRODUCTION (1000 tons)": "rice_production",
    
    "WHEAT AREA (1000 ha)": "wheat_area",
    "WHEAT PRODUCTION (1000 tons)": "wheat_production"
})
```

```python
df = df[[
    "state",
    "district",
    "year",
    "rice_area",
    "rice_production",
    "wheat_area",
    "wheat_production"
]]
```

```python
df.head()
```

```python
import xml.etree.ElementTree as ET

root = ET.Element("datasets")

for _, row in df.iterrows():
    record = ET.SubElement(root, "record")

    ET.SubElement(record, "state").text = str(row["state"])
    ET.SubElement(record, "district").text = str(row["district"])
    ET.SubElement(record, "year").text = str(row["year"])

    crops = ET.SubElement(record, "crops")

    # Rice
    rice = ET.SubElement(crops, "crop", name="rice")
    ET.SubElement(rice, "area").text = str(row["rice_area"])
    ET.SubElement(rice, "production").text = str(row["rice_production"])

    # Wheat
    wheat = ET.SubElement(crops, "crop", name="wheat")
    ET.SubElement(wheat, "area").text = str(row["wheat_area"])
    ET.SubElement(wheat, "production").text = str(row["wheat_production"])

tree = ET.ElementTree(root)
tree.write("data.xml")

print("XML file created successfully!")
```

```python
with open("data.xml", "r") as f:
    for i in range(10):
        line = f.readline()
        print(line[:200])   # only first 200 characters
```
